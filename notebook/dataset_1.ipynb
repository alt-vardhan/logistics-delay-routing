```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("E:\Projects\Supply Chain\Datasets\smart_logistics_dataset.csv")
print(df.shape)
df.head(20)
```

```python
df=df.drop(['Inventory_Level','Asset_ID','User_Transaction_Amount','User_Purchase_Frequency','Asset_Utilization'],axis=1)
df.head(10)
```

```python
df["Timestamp"] = pd.to_datetime(df["Timestamp"])

df["hour"] = df["Timestamp"].dt.hour
df["day"] = df["Timestamp"].dt.day
df["dayofweek"] = df["Timestamp"].dt.dayofweek
df["is_weekend"] = df["dayofweek"].isin([5, 6]).astype(int)
df=df.drop('Timestamp',axis=1)
df
```

```python
categorical_cols = ["Traffic_Status", "Shipment_Status"]
df_encoded = pd.get_dummies(df, columns=categorical_cols, drop_first=True)

X = df_encoded.drop(columns=["Logistics_Delay", "Logistics_Delay_Reason"],errors="ignore")
y = df_encoded["Logistics_Delay"]

df_encoded

```

```python
correlations = X.corrwith(y).sort_values(ascending=False)
correlations
```

```python
X = X.drop(columns=[c for c in X.columns if c.startswith("Shipment_Status")])
X
```

```python
X["congestion_index"] = (
    X["Traffic_Status_Heavy"] * 1.0 +
    X["Traffic_Status_Detour"] * 0.7 +
    (X["Waiting_Time"] / X["Waiting_Time"].max()) * 0.5 +
    (X["Demand_Forecast"] / X["Demand_Forecast"].max()) * 0.3
)
X
```

```python
X["is_peak_hour"] = X["hour"].isin([8,9,10,17,18,19]).astype(int)
X["peak_traffic_risk"] = X["is_peak_hour"] * (
    X["Traffic_Status_Heavy"] + X["Traffic_Status_Detour"]
)

X["lat_bin"] = (X["Latitude"] // 5).astype(int)
X["lon_bin"] = (X["Longitude"] // 5).astype(int)
X["geo_region"] = X["lat_bin"].astype(str) + "_" + X["lon_bin"].astype(str)
region_delay_rate = (
    pd.concat([X["geo_region"], y], axis=1)
    .groupby("geo_region")["Logistics_Delay"]
    .mean()
)

X["region_risk"] = X["geo_region"].map(region_delay_rate)
X["region_risk"].fillna(X["region_risk"].mean(), inplace=True)

X
```

```python
X["weather_stress"] = (
    (X["Temperature"] - X["Temperature"].mean()).abs() +
    (X["Humidity"] - X["Humidity"].mean()).abs()
)

hourly_demand = X.groupby("hour")["Demand_Forecast"].transform("mean")
X["relative_demand_pressure"] = X["Demand_Forecast"] / hourly_demand

X["delay_pressure"] = (
    X["Waiting_Time"] *
    (1 + X["Traffic_Status_Heavy"] + X["Traffic_Status_Detour"])
)

X = X.drop(columns=["lat_bin", "lon_bin", "geo_region"])
```

```python
correlations = X.corrwith(y).sort_values(ascending=False)
correlations
```

```python
X=X.drop(columns=["region_risk"])
X
```

```python
df_sorted = df.loc[X.index]

X_sorted = X.loc[df_sorted.index]
y_sorted = y.loc[df_sorted.index]

split_idx = int(0.8*len(X_sorted))

X_train = X_sorted.iloc[:split_idx]
X_test  = X_sorted.iloc[split_idx:]

y_train = y_sorted.iloc[:split_idx]
y_test  = y_sorted.iloc[split_idx:]
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

baseline = LogisticRegression(
    max_iter=1000,
    class_weight="balanced"
)

baseline.fit(X_train, y_train)

baseline_preds = baseline.predict(X_test)
print("BASELINE MODEL")
print(classification_report(y_test, baseline_preds))
```

```python
from xgboost import XGBClassifier

xgb_model = XGBClassifier(
    n_estimators=300,
    max_depth=6,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    eval_metric="logloss",
    random_state=42
)

xgb_model.fit(X_train, y_train)
```

```python
xgb_preds = xgb_model.predict(X_test)

print("XGBOOST MODEL")
print(classification_report(y_test, xgb_preds))
```

```python
y_probs = xgb_model.predict_proba(X_test)[:, 1]

from sklearn.metrics import precision_score, recall_score, f1_score

def evaluate_threshold(threshold):
    y_pred_thresh = (y_probs >= threshold).astype(int)
    return {
        "threshold": threshold,
        "precision": precision_score(y_test, y_pred_thresh),
        "recall": recall_score(y_test, y_pred_thresh),
        "f1": f1_score(y_test, y_pred_thresh)
    }

import pandas as pd

results = []
for t in [0.1, 0.2, 0.33, 0.4, 0.5]:
    results.append(evaluate_threshold(t))

pd.DataFrame(results)
```

```python
import lightgbm as lgb
from lightgbm import LGBMClassifier

lgb_model = LGBMClassifier(
    n_estimators=500,
    learning_rate=0.05,
    max_depth=-1,
    num_leaves=31,
    subsample=0.8,
    colsample_bytree=0.8,
    class_weight="balanced",
    random_state=42
)

lgb_model.fit(X_train, y_train)

```

```python
from sklearn.metrics import classification_report

lgb_preds = lgb_model.predict(X_test)

print("LIGHTGBM MODEL")
print(classification_report(y_test, lgb_preds))

```
